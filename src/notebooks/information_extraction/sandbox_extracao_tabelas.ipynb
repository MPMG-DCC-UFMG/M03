```python
table2df_func = lambda table: pd.DataFrame(table[1:], columns=table[0])

def count_table_empty(tables):
    cell_count = 0
    empty_count = 0
    for table in tables:
        for line in table:
            empty_count += line.count(None) + line.count('')
            cell_count += len(line)
    frac_empty = empty_count/cell_count if empty_count > 0 and cell_count > 0 else 0
    return frac_empty

def count_line_empty(line):
    return line.count(None) + line.count('')

def pdfplumber_extract_page(pdf, p, strategy):
    p_tables = pdf.pages[p].extract_tables({'snap_tolerance': 7, "horizontal_strategy": strategy})
    return {'page': p, 'tables': p_tables}

def camelot_extract_page(pdf, p):
    p_tables = camelot.read_pdf(file, flavor='stream', pages = str(p))
    p_tables =  list(map(lambda table: table.df, p_tables))
    return {'page': p, 'tables': p_tables}

def pdfplumber_default(file):
    pdf = pdfplumber.open(file)
    page_count = pdf.pages
    tables = []
    for p in range(page_count):
        p_tables = pdf.pages[p].extract_table({'snap_tolerance': 7})
        if bool(p_tables):
            tables.append({'page': p, 'tables': p_tables})
    return tables

def pdfplumber_horizontal_text(pdf):
    page = pdf.pages[p]
    tables = page.extract_tables({'horizontal_strategy': 'text'})
    return tables

def camelot_stream_flavor(file, p = 'all'):
    tables = camelot.read_pdf(file, flavor='stream', pages = p)
    return list(map(lambda table: table.df, tables))

def extract_id_from_path(file_path):
    id = file_path.split('/')[-1].split('_')[0]
    return id

cities_tables = {}

for city, files in cities_files_2.items():
    print(f'## {city} ##')
    cities_tables[city] = {}
    for file in files:
        file_id = extract_id_from_path(file)
        cities_tables[city][file_id] = []
        pdf = pdfplumber.open(file)
        for index in range(len(pdf.pages)):
            # extração usando texto
            tables_t = pdfplumber_extract_page(pdf, index, "text")
            # extração usando linha
            tables_l = pdfplumber_extract_page(pdf, index, "lines")
            # escolhe a tabela que possui menos células vazias
            tables = tables_t if count_table_empty(tables_t['tables']) < count_table_empty(tables_l['tables']) else tables_l
#             if 'c7d2371dbc67b51c8e350e03d0c46b8a' in file_id:
#                 print(tables)
            # se não encontrou nenhuma tabela, tenta extrair com o camelot
            if not bool(tables) or len(tables) == 0:
                print('** CAMELOT **')
                tables = camelot_extract_page(file, index + 1)
            if bool(tables) and bool(tables['tables']):
                # tabelas de uma linha ou coluna (TODO verificar se não é parte de uma tabela anterior ou subsequente)
                if len(tables['tables'][0]) ==1  or len(tables['tables'][0][0]) == 1:
                    print('**** desconsiderar ****')
                else:
                    cities_tables[city][file_id].append(tables)
        print('-'*50)
        
```

```python
def remove_none(line):
    new_line = []
    for content in line:
        if content != None:
            new_line.append(content)
    return new_line

def process_empty(line):
    processed_line = []
    empty_indexes = []
    for index in range(len(line)):
        content = line[index]
        if content != None:
            processed_line.append(content)
        if content == '':
            empty_indexes.append(index)
    return processed_line, empty_indexes

def process_line(index, lines):
    processed_info = process_empty(lines[index])
    #print('processed_info')
    #print(processed_info)
    processed_line, empty_indexes = processed_info[0], processed_info[1]
    #print('processed_line_1')
    #print(processed_line)
    #print('empty_indexes')
    #print(empty_indexes)    
    processed_line_empty_cells = count_line_empty(processed_line)
    #print('processed_line_empty_cells')
    #print(processed_line_empty_cells)
    # se não é a última linha
    if index < len(lines) - 1:
        # utiliza como referência 3 linhas abaixo
        reference_lines = [remove_none(line) for line in lines[index + 1:min(len(lines),index + 3)]]
        #print('reference_lines 1')
        #print(reference_lines)
    else:
        # utiliza como referência 3 linhas acima
        reference_lines = [remove_none(line) for line in lines[max(0,index - 3):index]]
        #print('reference_lines 2')
        #print(reference_lines)
    if len(reference_lines) == 0:
        reference_lines_size = 0
        reference_empty_lines = 0
    else: 
        reference_lines_size = int(sum([len(line) for line in reference_lines])/len(reference_lines))
        reference_empty_cells = sum([count_line_empty(line) for line in reference_lines])/len(reference_lines)
        #print('reference_lines_size')
        #print(reference_lines_size)
        #print('reference_empty_cells')
        #print(reference_empty_cells)
        #print('len(processed_line)')
        #print(len(processed_line))
    
    if len(processed_line) != reference_lines_size:
        new_line = [processed_line[index] for index in range(len(processed_line)) if index not in empty_indexes]
        if len(new_line) == reference_lines_size:
            processed_line = new_line
#         else:
#             #print('****** VERIFICAR ******')
    # alteração
    elif processed_line_empty_cells != reference_empty_cells and len(lines[index-1]) == len(processed_line) and (reference_empty_cells/len(processed_line)) >= 0.5:
            #print("processed_line_empty_cells{} != reference_empty_cells{} and len(lines[index-1]){} == len(processed_line){}".format(                processed_line_empty_cells, reference_empty_cells, len(lines[index-1]),len(processed_line)))
            previous_line = lines[index-1]
            processed_line = [f'{previous_line[i]} {processed_line[i]}' for i in range(len(previous_line))]
            lines[index-1][0] = REMOVAL_HASH
    return processed_line
```
